## Managing Data Frames with dplyr

`dplyr` verbs
- `select`: return a subset of the columns of the data frame
- `filter`: extract a subset of rows from a data frame wbased on logical conditions
- `arrange`: reorder rows of a dataframe
- `rename`: rename variables in a dataframe
- `mutate`: add new variables / columns or transform existing variables
- `summarize / summarise`: generate summary statistics of different variables in the data frame, possibly within strata

```r
library("dplyr")
```

```r
chicago <- mutate(chicago, 
       tempcat = factor(
           1 * (tmpd > 80), 
           labels = c("cold", "hot")
       )
      )
```

```r
hotcold <- group_by(chicago, tempcat)
head(hotcold, 10)
```

```r
summarise(hotcold, pm25 = mean(pm25), o3 = max(o3tmean2), no2 = median(no2tmean2))
```

```r
summarise(hotcold, pm25 = mean(pm25, na.rm = TRUE), o3 = max(o3tmean2), no2 = median(no2tmean2))
```

```r
chicago <- mutate(chicago, year = as.POSIXlt(date)$year + 1900)
years <- group_by (chicago, year)
```

```r
summarise(years, pm25 = mean(pm25, na.rm = TRUE), o3 = max(o3tmean2), no2 = median(no2tmean2))
```

```r
chicago %>% mutate(month = as.POSIXlt(date)$mon + 1) %>% group_by(month) %>% summarize(pm25 = mean(pm25, na.rm = T), o3 = max(o3tmean2), no2=median(no2tmean2))
```
