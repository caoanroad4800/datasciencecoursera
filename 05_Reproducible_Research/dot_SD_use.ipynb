```r
library(data.table)

DT = data.table(
    x = rep(c("a",'b','c'), each = 2)
    , y = c(1,3)
    , v = 1:6
)
setkey(DT, y)
DT
```

```r
DT[,
   .SD[, paste(x, v, sep="", collapse = "_"), by = y]
  ]
```
