## Task 4 - Prediction Model

The goal of this exercise is to build and evaluate your first predictive model. You will use the n-gram and backoff models you built in previous tasks to build and evaluate your predictive model. The goal is to make the model efficient and accurate.

**Tasks to accomplish**

- Build a predictive model based on the previous data modeling steps - you may combine the models in any way you think is appropriate.
- Evaluate the model for efficiency and accuracy - use timing software to evaluate the computational complexity of your model. Evaluate the model accuracy using different metrics like perplexity, accuracy at the first word, second word, and third word.

**Questions to consider**
- How does the model perform for different choices of the parameters and size of the model?
- How much does the model slow down for the performance you gain?
- Does perplexity correlate with the other measures of accuracy?
- Can you reduce the size of the model (number of parameters) without reducing performance?

### 1. Data Pre-processing

```r
# file path for english data
blogs_path <- "~/Soft/Rtest/JHU_capstone_project_data/final/en_US/en_US.blogs.txt"
news_path <- "~/Soft/Rtest/JHU_capstone_project_data/final/en_US/en_US.news.txt"
twitter_path <- "~/Soft/Rtest/JHU_capstone_project_data/final/en_US/en_US.twitter.txt"

# load data
blogs <- readLines(blogs_path, encoding = "UTF-8", skipNul = TRUE)
news <- readLines(news_path, encoding = "UTF-8", skipNul = TRUE)
twitter <- readLines(twitter_path, encoding = "UTF-8", skipNul = TRUE)

# Read the data files into data frames
blogs <- data.frame(text = blogs)
news <- data.frame(text = news)
twitter <- data.frame(text = twitter)
```

```r
# Sampling
set.seed(1565)
sample_pct <- 0.1

blogs_sample <- blogs %>% sample_n(., nrow(blogs)*sample_pct)
news_sample <- news %>% sample_n(., nrow(news)*sample_pct)
twitter_sample <- twitter %>% sample_n(., nrow(twitter)*sample_pct)

# Create aggregate sample
# agg_sample <- bind_rows(
#     mutate(blogs_sample, source = 'blogs')
#     , mutate(news_sample, source = 'news')
#     , mutate(twitter_sample, source = 'twitter')
# )

agg_sample <- bind_rows(
    mutate(blogs_sample)
    , mutate(news_sample)
    , mutate(twitter_sample)
)

# change agg_sample$source type: chr --> factor
# agg_sample$source <- factor(agg_sample$source)
```

```r
# Clean up to save RAM Space
rm(blogs
   , blogs_path
   , blogs_sample
   , news
   , news_path
   , news_sample
   , twitter
   , twitter_path
   , twitter_sample 
  )
```

```r
# to lower
agg_sample <- tolower(agg_sample)

# split at all "." "," and etc
agg_sample <- unlist(strsplit(agg_sample, "[.,:;!?(){}<>]+"))

# replace all non-alphanumeric characters with a space at the beginning/ending of a word
agg_sample <- gsub("^[^a-z0-9]+|[^a-z0-9]+$", " ", agg_sample) # at the begining/end of a line
agg_sample <- gsub("[^a-z0-9]+\\s", " ", agg_sample) # before space
agg_sample <- gsub("\\s[^a-z0-9]+", " ", agg_sample) # after space
agg_sample <- gsub("\\s+", " ", agg_sample) # remove multipe spaces
agg_sample <- str_trim(agg_sample) # remove spaces at the beginning / end of the line
```

```r
print(head(agg_sample, 20))
```

```r
# Remove spaces at the beginning / end of the line
agg_sample <- str_trim(agg_sample)

# Remove multiple spaces
agg_sample <- gsub("\\s+", " ", agg_sample)

# Remove blank lines
agg_sample <- agg_sample[nchar(agg_sample) > 0]

# save RDS with stop words
saveRDS(agg_sample, file = "~/Soft/Rtest/JHU_capstone_project_data/agg_sample_full.RDS")
```

```r
# Remove stop words
agg_sample <- unlist(lapply(agg_sample, function(x){removeWords(x, stopwords("en"))}))

# save RDS without stop words
saveRDS(agg_sample, file = "~/Soft/Rtest/JHU_capstone_project_data/agg_sample_nostopwords.RDS")
```

```r
# split words by space
words <- unlist(strsplit(agg_sample, "\\s+"))

# count word frequnce
word.freq <- table(words)

# convert to data frame
df <- cbind.data.frame(
    names(word.freq)
    ,as.integer(word.freq)
)
names(df) <- c("word", "freq")
row.names(df) <- df[,1]

# sort words by frequence decending
df <- df[order(-df$freq), ]

# save as RDS file
saveRDS(df, file = "~/Soft/Rtest/JHU_capstone_project_data/co_1gram_en.RDS")
```

### Unique words needed to cover the text

```r
df <- readRDS("~/Soft/Rtest/JHU_capstone_project_data/co_1gram_en.RDS") # with stop words
```

```r
df$count <- 1
df$count <- cumsum(df$count)
df$coverage <- cumsum(df$freq) / sum(df$freq) * 100
df <- df[df$coverage <= 91, ]

# find the word counts for 50% and 90% coverage

points <- rbind(
    tail(df[df$coverage <= 50,], 1)
    ,tail(df[df$coverage <= 90,], 1)
)
```

```r
g <- NULL
g <- ggplot(data = df, aes(x = count, y = coverage, group = 1)) +
geom_line() +
geom_point(data = points, colour = "red", size = 3) +
geom_text(data = points, aes(label = count), hjust = -1, vjust = 1) +
ggtitle("Number of Words to Cover Text") + 
xlab("Number of Words") + 
ylab("Coverage %")
g
```
